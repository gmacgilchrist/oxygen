## Watermass transformation in oxygen space

```python
import xarray as xr
import cftime
import utils.geo as geo
from xhistogram.xarray import histogram
import numpy as np
from dask.diagnostics import ProgressBar
import matplotlib.pyplot as plt
```

```python
rootdir = '/projects/SOCCOM/datasets/gobai-o2/'
filename = 'GOBAI-O2-v1.0_processed.zarr'
ds = xr.open_zarr(rootdir+filename)
grid = xr.open_dataset(rootdir+'regular_grid.nc')
```

```python
grid['volcello']=grid['rC']*grid['dp']
```

```python
oxy_bins = np.array(np.linspace(0,400,400))
dovals = np.diff(oxy_bins)
```

```python
hs = histogram(ds['oxy'],bins=[oxy_bins],weights=grid['volcello'],dim=['lon','lat','pres'])
# Replace oxy_bin coordinate with upper bound of bin
hs = hs.assign_coords({'oxy_bin':oxy_bins[1:]})
# Normalize each bin by bin width
doxy = xr.DataArray(dovals,dims='oxy_bin',coords={'oxy_bin':hs['oxy_bin']})
hs = hs/doxy
```

```python
with ProgressBar():
    hs = hs.compute()
```

```python
t1 = ['2005','2009']
t2 = ['2015','2019']
t1slice = slice(t1[0]+'-01-01',t1[1]+'-12-31')
t2slice = slice(t2[0]+'-01-01',t2[1]+'-12-31')
t1label = t1[0]+' to '+t1[1]
t2label = t2[0]+' to '+t2[1]
t1color='tab:blue'
t2color='tab:orange'
dt = cftime.datetime(int(t2[1]),12,31)-cftime.datetime(int(t1[1]),12,31)
dt = dt.days*24*60*60
```

```python
fig,ax = plt.subplots(figsize=(12,12),nrows=3)

# Calculations
hs_t1 = hs.sel(time=t1slice).mean('time')
hs_t2 = hs.sel(time=t2slice).mean('time')
wm_t1 = hs_t1['oxy_bin'].weighted(hs_t1.fillna(0)).mean('oxy_bin')
wm_t2 = hs_t2['oxy_bin'].weighted(hs_t2.fillna(0)).mean('oxy_bin')
hs_diff = (hs_t2-hs_t1)/dt
x = (hs_diff*doxy).values
hs_trans = np.nancumsum(x[::-1])[::-1] 

# Plot
ax[0].plot(hs_t1['oxy_bin'],hs_t1,label=t1label,color=t1color)
ax[0].plot(hs_t2['oxy_bin'],hs_t2,label=t2label,color=t2color)
ax[0].text(300,3.5e15,'Volume-weighted mean oxygen :',fontweight='bold')
ax[0].text(300,3e15,str(np.round(wm_t1.values,2)),color=t1color,fontweight='bold')
ax[0].text(300,2.5e15,str(np.round(wm_t2.values,2)),color=t2color,fontweight='bold')
ax[0].set_ylabel('VOLUME [m3 / umol kg-1]')
ax[0].set_xlabel('OXYGEN [umol kg-1]')
ax[0].set_title('OCEAN VOLUME IN OXYGEN SPACE',fontsize=14)
ax[0].legend(loc='upper left',frameon=False)

# doxy = hs['oxy_bin'][1]-hs['oxy_bin'][0]
ax[1].bar(hs_diff['oxy_bin'],hs_diff,width=doxy)
ax[1].set_ylabel('VOLUME DIVERGENCE [m3 s-1 / umol kg-1]')
ax[1].set_xlabel('OXYGEN [umol kg-1]')
ax[1].set_title('RATE OF VOLUME CHANGE BETWEEN '+t2label+' AND '+t1label,fontsize=14)
# ax[1].plot(hs_diff['oxy_bin'],hs_trans)

# ax[2]
ax[2].plot(hs_diff['oxy_bin'],hs_trans)
ax[2].set_ylabel('VOLUME DIVERGENCE [m3 s-1]')
ax[2].set_xlabel('OXYGEN [umol kg-1]')
ax[2].set_title('VOLUME TRANSFORMATION BETWEEN '+t2label+' AND '+t1label,fontsize=14)

fig.tight_layout()
fig.savefig('../figures/GOBAI-O2/oxygen-transformation_gobai.png',dpi=300)
```
