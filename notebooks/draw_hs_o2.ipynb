```python
import xarray as xr
```

```python
config_id = 'ESM4_historical_D1'
basin = 'global'
hs_variables = ['o2']
weightname = 'volcello'
hs_datadir = '../data/histograms/'
hs_filename = 'hs.'+config_id+'.'+'-'.join(hs_variables)+'.'+weightname+'.'+basin+'.nc'

hs = xr.open_dataset(hs_datadir+hs_filename)
```

```python
hs['volcello'].isel(time=slice(0,20)).mean('time').plot()
hs['volcello'].isel(time=slice(-20,-1)).mean('time').plot()
```

```python
ds_datadir = '/work/gam/projects/oxygen/data/'
ds_filename = 'ds.'+config_id+'.zarr'
ds = xr.open_zarr(ds_datadir+ds_filename)
```

```python
(ds['o2'].isel(time=slice(-20,-1)).mean('time').sum('z_l')-ds['o2'].isel(time=slice(0,20)).mean('time').sum('z_l')).plot(robust=True)
```

```python
(hs['volcello'].isel(time=slice(-20,-1)).mean('time')-hs['volcello'].isel(time=slice(0,20)).mean('time')).cumsum('o2_bin').plot()
```
